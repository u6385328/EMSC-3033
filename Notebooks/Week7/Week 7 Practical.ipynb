```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = 23065.85 # Given cartesian coordinates
Y = 5202.38 
Z = -52955.11

lat = -35.32
long = 149.36
```

### Part 1 Questions
#### Part A

```python
#Basic cartesian conversion

B = np.sqrt(X**2+Y**2+Z**2)
print("The total field vector, B, is equal to "+ str(np.around(B,1)) +" nT")

H = np.sqrt(X**2 + Y**2)
print("The horizontal component, H, is equal to "+ str(np.around(H,1)) +" nT")

I = np.rad2deg(np.arctan(Z/H))
print("Inclination, I, is equal to "+ str(np.around(I,1)) +" degrees")

D = np.rad2deg(np.arctan2(Y,X))
print("Declination, D, is equal to "+ str(np.around(D,1)) +" degrees")
```

```python
#GAD Conversion
I_GAD = np.rad2deg(np.arctan(2*np.tan(np.deg2rad(lat))))
print("GAD Inclination, I, is equal to "+ str(np.around(I_GAD,1)) +" degrees")

D_GAD = 0
print("GAD Declination, D, is always equal to "+ str(np.around(D_GAD,1)) +" degrees")
```

### Part 2 Questions
#### Part A

```python
B = 50000E-9 # Earth's magnetic field strength, nT 
A = 0.5 # Area, m^2
k = 0.025 # Magnetic susceptibility
Z_p = 5 # Depth of monopole, m
s = 0.1 # Anomoly step size, m
xmin = -15 # m
xmax = 15.1 # m
I = np.deg2rad(45) # Inclination, degrees
m = B * k * A # Magnetic moment
```

```python
x = np.arange(-15, 15.1, 0.1)
r = np.sqrt(x**2 + Z_p**2)
```

```python
delta_Z = -((m*Z_p)/(r**3)) # Vertical 

delta_H = (m*x)/(r**3) # Horizontal

delta_B = (m/(r**3))*(x*np.cos(I) - Z_p*np.sin(I)) # Total

#print(delta_Z, delta_H, delta_B)
```

```python
plt.plot(x, delta_Z*1E9, 'r', label='Vertical')
plt.plot(x, delta_H*1E9, 'g', label='Horizontal')
plt.plot(x, delta_B*1E9, 'b', label='Total')
plt.legend()
plt.ylabel("Anomoly Strength (nT)")
plt.xlabel("x (m)")
plt.title("Anomoly Strength with I = 45 Degrees")
plt.show()
```
